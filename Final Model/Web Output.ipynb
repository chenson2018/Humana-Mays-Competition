```python
import pandas as pd
```

```python
training = pd.read_csv('HMAHCC_COMP.csv')
test = pd.read_csv('HMAHCC_HOLDOUT.csv')
```

```python
test = test[['ID', 'EVENT_DESCR', 'EVENT_ATTR1', 'EVENT_ATTR2', 'EVENT_ATTR3',
       'EVENT_ATTR4', 'EVENT_ATTR5', 'EVENT_ATTR6', 'EVENT_ATTR7',
       'EVENT_ATTR8', 'EVENT_ATTR9', 'EVENT_ATTR10','days', 'PAY_DAY_SUPPLY_CNT',
       'PAYABLE_QTY', 'MME', 'DRUG_TYPE', 'SPECIALTY', 'SPECIALTY2',
       'SPECIALTY3']]

test.columns = training.columns
```

```python
from humana_class import DataPrep
```

```python
training_class = DataPrep(test)
```

```python
training_class.seperate_frame()
```

```python
patient1 = 'ID38179240346'
patient2 = 'ID14418595663'
patient3 = 'ID4467853207'
```

```python
def return_frames(patient_id):
    
    diag = training_class.new_diagnosis[training_class.new_diagnosis.id == patient_id]
    reject = training_class.rx_reject[training_class.rx_reject.id == patient_id]
    paid = training_class.rx_paid[training_class.rx_paid.id == patient_id]
    prov = training_class.new_provider[training_class.new_provider.id == patient_id]
    call = training_class.call[training_class.call.id == patient_id]
    non_rx = training_class.non_rx_claim[training_class.non_rx_claim.id == patient_id]
    
    l = [diag, reject, paid, prov, call, non_rx]
    
    return l
```

```python
p1_frames = return_frames('ID14418595663')
p2_frames = return_frames(patient2)
p3_frames = return_frames(patient3)
```
